```python
def normalize_series(seriesData):
    _max = seriesData.max()
    _min = seriesData.min()
    return seriesData.map(lambda x: (x - _min) / (_max - _min))


titanicData.PassengerId = normalize_series(titanicData.PassengerId)
titanicData.Age = normalize_series(titanicData.Age)
titanicData.Fare = normalize_series(titanicData.Fare)
titanicData.Pclass = normalize_series(titanicData.Pclass)
titanicData.Sex = normalize_series(titanicData.Sex)
titanicData.Embarked = normalize_series(titanicData.Embarked)
titanicData.SibSp = normalize_series(titanicData.SibSp)
titanicData.Parch = normalize_series(titanicData.Parch)

titanicData.describe()
```

```python
import pandas as pd
import math


def GetCount(data: pd.Series, value) -> int:
    return data[data == value].count()



data = {
    "Age": [25, 30, 22, 35, 40, 50, 28, 45],
    "Income": ["High", "High", "Medium", "Low", "Medium", "High", "Low", "Low"],
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Male"],
    "Bought": ["Yes", "No", "Yes", "No", "Yes", "No", "Yes", "No"],
}
df = pd.DataFrame(data)
df.Age = df.Age.map(lambda x: "Young" if x < 30 else "Old")

p_Yes = GetCount(df.Bought, "Yes") / len(df)
p_No = GetCount(df.Bought, "No") / len(df)
entropy = -(p_Yes * math.log2(p_Yes) + p_No * math.log2(p_No))
print(f"p_Yes:{p_Yes} p_No:{p_No} Entropy:{entropy}")

p_Young_Yes = GetCount(df[df.Age == "Young"].Bought, "Yes") / len(df[df.Age == "Young"])
p_Young_No = GetCount(df[df.Age == "Young"].Bought, "No") / len(df[df.Age == "Young"])

p_Old_Yes = GetCount(df[df.Age == "Old"].Bought, "Yes") / len(df[df.Age == "Old"])
p_Old_No = GetCount(df[df.Age == "Old"].Bought, "No") / len(df[df.Age == "Old"])

entropy_Young = -(p_Young_Yes * math.log2(p_Young_Yes) + p_Young_No * math.log2(p_Young_No))
entropy_Old = -(p_Old_Yes * math.log2(p_Old_Yes) + p_Old_No * math.log2(p_Old_No))
print(f"p_Young_Yes:{p_Young_Yes} p_Young_No:{p_Young_No} Entropy:{entropy_Young}")
print(f"p_Old_Yes:{p_Old_Yes} p_Old_No:{p_Old_No} Entropy:{entropy_Old}")
```

```python
def AddPadding(image,padLeft,padRight,padTop,padBottom,padNo=0):
    newImage = []
    for i in range(padTop):
        newImage.append([padNo]*(len(image[0])+padLeft+padRight))
    for i in range(len(image)):
        newImage.append([padNo]*padLeft+image[i]+[padNo]*padRight)
    for i in range(padBottom):
        newImage.append([padNo]*(len(image[0])+padLeft+padRight))
    return newImage

def Convolve(image, kernel,stride):
    kernel_height = len(kernel)
    kernel_width = len(kernel[0])
    image_height = len(image)
    image_width = len(image[0])
    result = []
    for i in range(0,image_height-kernel_height+1,stride):
        row = []
        for j in range(0,image_width-kernel_width+1,stride):
            sum = 0
            for k in range(kernel_height):
                for l in range(kernel_width):
                    sum += image[i + k][j + l] * kernel[k][l]
            row.append(sum)
        result.append(row)
    return result

def MaxPooling(image,kernal):
    kernel_height = len(kernal)
    kernel_width = len(kernal[0])
    image_height = len(image)
    image_width = len(image[0])
    result = []
    for i in range(0,image_height-kernel_height+1,1):
        row = []
        for j in range(0,image_width-kernel_width+1,1):
            max = -1
            for k in range(kernel_height):
                for l in range(kernel_width):
                    if image[i + k][j + l] > max:
                        max = image[i + k][j + l]
            row.append(max)
        result.append(row)
    return result

image = [
    [3,3,2,1,0],
    [0,0,1,3,1],
    [3,1,2,2,3],
    [2,0,0,2,2],
    [2,0,0,0,1]
]
kernel = [
    [0,1,2],
    [2,2,0],
    [0,1,2]
]
paddedImage=AddPadding(image,1,1,1,1,0)
print(f"paddedImage:\n{paddedImage}")
print(Convolve(paddedImage, kernel,1))
print(Convolve(image, kernel,2))
```
